```python
import tensorflow as tf
import numpy as np
import t3f
import random
def TTL(x, reuse):
    # A 784 x 625 TT-matrix.
    with tf.variable_scope(tf.get_variable_scope(), reuse=True): 
#         matrix_shape = ((4, 7, 4, 7), (5, 5, 5, 5))
        matrix_shape = ((4, 7, 4, 7), (5, 5, 5, 5))
        tt_W_1 = t3f.get_variable('tt_W_1', initializer=t3f.random_matrix(matrix_shape))
        h = tf.nn.relu(t3f.matmul(tt_W_1, x))
    return h
```

    Table 1: A numerical example of compressing with TT-RNNs. Assuming that an input dimension of 160×120×3 is
factorized as 8 × 20 × 20 × 18 and the hidden layer as 4 × 4 × 4 × 4 = 256, depending on the TT-ranks we calculate the
number of parameters necessary for a Fully-Connected (FC) layer, a TTL which is equivalent to TT-SRNN , TT-LSTM and
TT-GRU in their respective vanilla and parallelized form

```python
import tensorflow as tf
import numpy as np
import t3f
import random

input_shape = (160,120,3)
input_dim = input_shape[0]*input_shape[1]*input_shape[2]
print(input_dim)
hidden_dim = 256
input_tt_shape = (8, 20, 20, 18)
hidden_tt_shape = (4, 4, 4, 4)
with tf.variable_scope("ansu", reuse=True): 
    x = tf.placeholder(tf.float32, shape=(input_dim,1))
    h_prev = tf.placeholder(tf.float32, shape=(hidden_dim, 1))
    
    W_shape = (hidden_tt_shape, input_tt_shape)
    U_shape = (hidden_dim, hidden_dim)
    
    W_r = t3f.get_variable('W_r', initializer=t3f.random_matrix(W_shape))    
    U_r = tf.get_variable('U_r', U_shape)  
    
    W_z = t3f.get_variable('W_z', initializer=t3f.random_matrix(W_shape))    
    U_z = tf.get_variable('U_z', U_shape) 
    
    W_d = t3f.get_variable('W_d', initializer=t3f.random_matrix(W_shape))    
    U_d = tf.get_variable('U_d', U_shape)

    r = tf.nn.sigmoid(t3f.matmul(W_r, x) + tf.matmul(U_r, h_prev))
    z = tf.nn.sigmoid(t3f.matmul(W_z, x) + tf.matmul(U_z, h_prev))
    d = tf.nn.tanh(t3f.matmul(W_d, x) + tf.matmul(U_d, (r*h_prev)))
    h = (np.ones((hidden_dim,1))-z)*d + z*h_prev

    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())   
        outt = sess.run(h, feed_dict={x: np.random.rand(input_dim,1), h_prev : np.random.rand(hidden_dim,1)})
        print(r.shape,z.shape,d.shape,h.shape)
        for i in range(100):
            outt = sess.run(h, feed_dict={x: np.random.rand(input_dim,1), h_prev : outt})
        print(result)
```
